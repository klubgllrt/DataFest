```
# Extract the regressand, Y, from the data frame
sampleData_Y = df.loc[:, 'Fatigue']
sampleData_Y.head()
```

```
#Extract the regressors, X, from the data frame

sampleData_X = df.loc[:, ('TeamPoints', 'TeamPointsAllowed', 'Distance Ran (m)', 'Fatigue', 'Soreness', 'Desire', 'Irritability', 'SleepHours', 'SleepQuality', 'MonitoringScore', 'TrainingReadiness', 'Games_LW', 'Flight')]
sampleData_X.columns
sampleData_X.head()
```

```
# determine the significance of feature.
from sklearn.feature_selection import SelectKBest, f_regression
k=2
selector = SelectKBest(f_regression, k=k)
selector.fit(sampleData_X, sampleData_Y)
best_feats = selector.get_support(indices = True)
xTrain_best = sampleData_X.iloc[:, best_feats]
xTrain_best.head()
```
